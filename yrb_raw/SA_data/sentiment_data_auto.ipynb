## Create a DataFrame with All the Results

Create a function called `all_results_df(news, auto)` that will transform the `articles` list in 
a DataFrame. The results will have two parameters: `news` is the articles' list and 
`auto` is a string to specify the type of automobile company listed of the news articles:
Ford, GM, or Tesla.

The resulting DataFrame should have the following columns:

* Title: The article's title
* Description: The article's description
* Text: The article's content
* Date: The date when the article was published on the format `YYY-MM-DD` (eg. 2019-07-11)
* Auto: A string specifying the type of auto manufacturer that was in the news article
(`ford` for Ford, `gm` for GM, and 'tesla' for Tesla)

def all_results_df(news, auto):
    articles = []
    for article in news:
        try:
            title = article["title"]
            description = article["description"]
            text = article["content"]
            date = article["publishedAt"][:10]
            
            articles.append({
                "title": title,
                "description": description,
                "text": text,
                "date": date,
                "auto": auto
            })
        except AttributeError:
            pass

    return pd.DataFrame(articles)
